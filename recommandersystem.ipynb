# Part 2: Building a simple recommender

```python
small_df2 = movie_df.loc[:, ["title", "runtime", "vote_count", "vote_average"]]
```

```python
small_df2.head(10)
```

```python
# Prerequisite 1: Only considering movies that are greater than 45 minutes and less than 300 minutes in length.

new_df = small_df2[(small_df2['runtime'] > 45) & (small_df2['runtime'] < 300)]
```

```python
new_df['runtime'].describe() 
```

```python
new_df.head(10)
```

```python
new_df['vote_count'].describe()
```

```python
#Prerequisite 2:  
#For a movie to be considered in the rankings, it must have garnered more votes than at least 80% of the movies present in raw dataset
vote_req = new_df['vote_count'].quantile(0.8) 
movie_chart = new_df[new_df['vote_count'] > vote_req]

movie_chart
```

```python
m = vote_req + 1 #minimum number of votes required for the movie to be in the chart
#print(m)
c = movie_chart["vote_average"].mean() #the mean rating of all the movies in the dataset
#print(c)
```

# 1. Calculate the weighted rating score.
# 2. Add the score as a feature to the dataframe, by using Apply function. 

```python
def weighted_avg(x):
  wr = ((x.vote_count * x.vote_average) + (m * c))/(x.vote_count +m)
  return wr
```

```python
movie_chart["score"] = movie_chart.apply(weighted_avg, axis = 1 )
movie_chart
```

#3. Sort the dataframe on the basis of the score just computed.
#4. Output of the list of top movies. 

```python
sorted_df = movie_chart.sort_values('score', ascending=False).head(10)
sorted_df.loc[:,["title","runtime","vote_average","score"]]
```

# Part 3:  Implement a user-based collaborative filtering system

#1. Using the data in ratings_small.csv, construct the user rating matrix (where the rows represent the users, and the columns the movies).

```python
df_ratings = pd.read_csv("ratings_small.csv")
```

```python
df_ratings.head(10)
```

```python
user_ratingmatrix = df_ratings.pivot_table(index = "userId", columns = "movieId", values = "rating")
```

```python
user_ratingmatrix #matrix showing the user preferences 
```

# 2. This matrix (user_ratingmatrix) will be sparse (lot of unknown values), explain why. Propose a way to fill in the missing values and justify it. Implement it to have a matrix without missing values.

***Answer:*** The matrix rows represents the users and the columns represents the movies. If a particular movie has been rated by a particular user, only then the value appers in that particular position in the matrix. Since there are 9066 columns i.e, there are 9066 movies, for a user it's kinda unlikely that all movies to be rated. Additionally it's unlikely to expect that a particular movie to be rated by all users. That's why there are a lot of unknown values in the sparse matrix.

We can think of filling up the unknown values by 0. Because we don't really know the values that a user could rate for particular movie. If we take the approach to fill up the known values of the movies by the average of all ratings given by the user or by the average rating of a movie that can be misleading and biased. 

# 3. Calculate the cosine similarity for each pair of users, in order to obtain a user-user similarity matrix.

```python
from sklearn.metrics.pairwise import cosine_similarity
cosinesim_user = cosine_similarity(user_ratingmatrix.fillna(0))
np.fill_diagonal(cosinesim_user, 0) # the diagonals values signifies the same user for which the cosine sim. value is 1, and it's not necessary for the calculation. Hence we make it 0. 
df_cosinesim_user = pd.DataFrame(cosinesim_user, index = user_ratingmatrix.index, columns=user_ratingmatrix.index)
```

```python
df_cosinesim_user #user-user similarity matrix
```

```python
h = df_cosinesim_user.loc[1,:].sort_values(axis = 0, ascending=False)
print(h.head(10)) #Top 10 users who are similar to the user with userId: 1. 
```

# 4.  Implement a function to return the n nearest neighbors of a given user in the sense of the calculated similarity.

```python
def n_neighbor(k, n):       #k is the user for whom we are calculating the n number of nearest neighbors.
  neighbors = df_cosinesim_user.loc[k,:].sort_values(axis = 0, ascending=False)
  return neighbors.head(n)

call1 = n_neighbor(1,5)

print("5 nearest neighbors of the user_id= 1:",list(call1.index))
```

# 5. Using the preceding function, implement a function that recommends k movies—at most—for a given user. Explain how this function works.


```python
def score_function(x): #It is the score function to calculate the weighted movie scores of the movies reviewed by similar user
  s = []
  d = []
  score = []
  l = np.unique(list(x.movieId))
  for i in l: 
    s = list(x[x["movieId"]==i]["movie_id_cos_mul"]) #obtiaining all the (cosinesimilarity * movie rating) of a particular movie
    sum_s = sum(s)
    d = list(x[x["movieId"]==i]["cosine_value"]) #obtiaining the cosine similarity of all users who has rated that particular movie
    sum_d = sum(d)
    if sum(d) > 0.0 and sum(s) > 0.0:
      score.append(sum_s/sum_d) #adding the weighted score to a list
    else:
      score.append(0)
  return score
```

```python
def movie_recommender(n,k,n_nearest = 8): #for userId n we are calculating the  k movie recommendations.

  similar_userton = n_neighbor(n,n_nearest) # By default we will be calculating the 8 neartest neighbor to n.
  dict_similar_userton = similar_userton.to_dict() #Converting similar_userton  to a dictionary 
  recommendar_group = df_ratings[df_ratings['userId'].isin(list(similar_userton.index))] #data frame containing the movie rating of sim. users.
  
  user_n = df_ratings[df_ratings["userId"].isin([n])] #movie ratings of the user with the userId n
  a = list(user_n["movieId"]) # making a list that has already been rated by the userId n

  #filtered_recommendar_group is the filtered dataframe containing the movies that aren't yet watched by userId n
  filtered_recommendar_group = recommendar_group[~recommendar_group["movieId"].isin(a)]

  #Mapping the cosine values of the users corresponding to the userIds 
  filtered_recommendar_group["cosine_value"] = filtered_recommendar_group["userId"].map(dict_similar_userton)

  # Multiplying the cosine similarity of a particular user with the all movies that has been watched by the user
  filtered_recommendar_group["movie_id_cos_mul"] = filtered_recommendar_group["cosine_value"] * filtered_recommendar_group["rating"]
  #sorting the filtered_recommendar_group accoring to the movie Id's 
  sorted_filtered_rg = filtered_recommendar_group.sort_values('movieId')
  smallsorted_filtered = sorted_filtered_rg.loc[:, ["movieId", "cosine_value", "movie_id_cos_mul"]]

  #Calling the score_function that was defined before movie_recommender() to calculate the weighted scores of the movies that are going to be recommeded on the basis of the user similarity.
  score_list = score_function(smallsorted_filtered)
  unique_movieIds = np.unique(list(smallsorted_filtered.movieId))

  #Creating a dataframe containing all the movies along with their weighted scores 
  rec_movies = pd.DataFrame(list(zip(unique_movieIds, score_list)), columns = ["movieId", "score"])

  #Creating a dataframe containg the k number of movieIds with the heighest scores 
  sorted_rec_movies = rec_movies.sort_values("score", ascending=False).head(k)
  return sorted_rec_movies



c = movie_recommender(2,10)
print("Reommended movie list for userId 2 is: \n ", c)
```
